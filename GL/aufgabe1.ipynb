```python
import numpy as np
import matplotlib.pyplot as plt

from constants import FIGSIZE
```

```python
# Define resonator parameters
g_1 = 1
g_2 = 0.5
```

```python
# Define the border of the stable area
PLOT_LIMIT = 3

g_1_array = np.linspace(-PLOT_LIMIT, PLOT_LIMIT, 200)
g_2_array = 1 / g_1_array
```

```python
# Plot the stability diagram

fig, ax = plt.subplots(figsize=FIGSIZE)
ax.grid()
ax.set_xlabel('$g_1$')
ax.set_ylabel('$g_2$')

ax.fill_between(g_1_array, np.zeros(g_1_array.shape), g_2_array,
                hatch='/', facecolor=(0, 0, 0, 0), edgecolor='b',  label='Stabiler Bereich')
ax.plot(g_1, g_2, 'kx', label='Vorliegender Resonator')

ax.set_xlim(-PLOT_LIMIT, PLOT_LIMIT)
ax.set_ylim(-PLOT_LIMIT, PLOT_LIMIT)

ax.legend()

fig.savefig('figs/aufgabe_1.svg', bbox_inches='tight')
```
