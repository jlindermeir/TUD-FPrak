```python
import numpy as np
import matplotlib.pyplot as plt

from constants import FIGSIZE
```

```python
# Powermeter data ([m], [W], [W])
L_arr, P_arr, std_P_arr = np.loadtxt(
    fname='data/Powermeter/Messwerttabellen - Resonatorlänge.csv',
    delimiter=',',
    skiprows=1
).T

# Length measurement error [m]
std_L = 0.5e-2
```

```python
# Create plot
fig, ax = plt.subplots(figsize=FIGSIZE)
ax.set_xlabel('Resonatorlänge L [m]')
ax.set_ylabel('Ausgangleistung P [mW]')
ax.grid()

# Plot measurements, convert power values to mW
ax.errorbar(x=L_arr, xerr=std_L, y=P_arr * 1e3, yerr=std_P_arr * 1e3,
            fmt='kx-')
```

```python
fig.savefig('figs/aufgabe3a.svg', bbox_inches='tight')
```
