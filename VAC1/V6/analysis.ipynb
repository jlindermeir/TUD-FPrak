```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.optimize import curve_fit
```

```python
filename = 'data/p_t-csv'

t_arr, p_H_arr = np.loadtxt(
    fname=filename,
    delimiter=',',
    unpack=True,
    skiprows=1
)
```

```python
p_fit = lambda t, A, B, C: A * np.exp(-B * t) + C

params_0 = [
    p_H_arr[0],
    1,
    0
]

params_fit, _ = curve_fit(p_fit, t_arr, p_H_arr, p0=params_0)

print(params_fit)
```

```python
fig, ax = plt.subplots(figsize=[10,10])

ax.plot(t_arr, p_H_arr, 'k+')

ax.plot(t_arr, p_fit(t_arr, *params_fit))

print(p_H_arr.max())
```
