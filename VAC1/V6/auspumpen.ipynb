```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load measurements
filenames = [
    ('Heizung an', 'data/Messwerte Richard VAK1 - Auspumpen Turbomolekular, Heizung an.csv'),
    ('Heizung aus', 'data/Messwerte Richard VAK1 - Auspumpen Turbomolekular, Heizung aus.csv')
]

dataset = {
    name: np.loadtxt(
        fname=filename,
        delimiter=',',
        unpack=True,
        skiprows=1
    )
    for name, filename in filenames
}
```

```python
width_ratios = [50, 20]

# Plot measurements
fig, axes = plt.subplots(1, 2, figsize=[10,8], sharey=True, gridspec_kw={'width_ratios': width_ratios})
for ax, (name, arrays), ratio in zip(axes, dataset.items(), width_ratios):
    ax.grid()
    ax.set_xlabel('$t$ [min]')
    ax.set_xlim([0, ratio])
    ax.set_ylabel('$p_H$ [mbar]')
    ax.set_yscale('log')

    ax.plot(
        arrays[0] / 60,  # Convert to minutes
        arrays[1],
        'kx',
        label=name
    )
    ax.legend()

    print(arrays[0][np.argmax(arrays[1])]/60)

fig.savefig('figs/auspumpen.pdf', bbox_inches='tight')
```
